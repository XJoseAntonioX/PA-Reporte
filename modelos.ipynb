# Choose best combination hyperparameters and parameters

```python
def best_combinations(feature_names, diccionario_municipios, degree_search):
    
    kf = KFold(n_splits=20)

    for municipio, df in diccionario_municipios.items():

        Y = df['Dato Sinaica']

        mean_mae = []
        variables = []
        degree_model = []

        for r in range(1, len(feature_names) + 1):
                for combo in combinations(feature_names, r):
                        selected_indices = list(combo)
                        X = df.loc[:, selected_indices]
                        score, param = Hyperparameter(X,Y, degree_search)
                        poly = PolynomialFeatures(degree = param["poly__degree"], 
                                                        include_bias = param["poly__include_bias"], 
                                                        interaction_only = param["poly__interaction_only"], 
                                                        order = param["poly__order"])

                        X_selected = poly.fit_transform(X)

                        scores_each_fold = []

                        for i, (train_index, test_index) in enumerate(kf.split(X_selected)):
                                X_train = X_selected[train_index]
                                Y_train = Y.iloc[train_index]
                                X_test = X_selected[test_index]
                                Y_test = Y.iloc[test_index]

                                model = sm.OLS(Y_train, X_train).fit()
                                predictedY = model.predict(X_test)

                                scores_each_fold.append(mean_absolute_error(Y_test, predictedY))
                        
                        degree_model.append(param["poly__degree"])
                        mean_mae.append(np.mean(scores_each_fold))
                        variables.append(tuple(selected_indices))
                        
        results_df = pd.DataFrame({"Errores Modelo MAE (µg/m3)": mean_mae, "Grado Del Modelo": degree_model}, index=variables)
        results_df = results_df.sort_values(by="Errores Modelo MAE (µg/m3)")

        print(f"Los resultados para el municipio de {municipio} son:")
        display(results_df.iloc[:6])
```

# Hyperparameter tuning

```python
def Hyperparameter(X,Y, degree):
    MAE = make_scorer(mean_absolute_error, greater_is_better = False)

    pipeline = Pipeline([
        ('poly', PolynomialFeatures()),
        ('linear', LinearRegression())
        ])

    param_grid = [
        {"poly__degree": degree,
        'poly__interaction_only': [True, False],
        "poly__include_bias": [True, False],
        "poly__order": ["C", "F"]}
    ] 

    grid_search = GridSearchCV(pipeline, param_grid, cv=20, scoring=MAE)
    grid_search.fit(X, Y)

    return (grid_search.score(X,Y), grid_search.best_params_)
```
